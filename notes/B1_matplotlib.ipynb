ベクトルの内積

$\vec{a}\cdot\vec{b} = (a_1,a_2,\cdots,a_n)\cdot(b_1,b_2,\cdots,b_n)$

$= (a_1 b_1 + a_2 b_2 + \cdots + a_n b_n)$

$= \sum_{k=1}^n a_k b_k$

```python
# ベクトルの内積（各要素同士の積の総和）
# 2つのベクトルの要素数が同じである必要がある
import numpy as np
a = np.array([1, 2, 3])
b = np.array([3, 2, 1])
print("--- dot()関数 ---")
print(np.dot(a, b)) # dot()関数による内積
print("--- 積の総和 ---")
print(np.sum(a * b)) # 積の総和による内積
```

ノルム

$L^2 ノルム$

$\lVert\vec{x}\rVert_2 = \sqrt{x_1^2+x_2^2+\cdots+x_n^2}$

$=\sqrt{\sum_{k=1}^n x_k^2}$

$L^1 ノルム$

$\lVert\vec{x}\rVert_1 = |x_1|+|x_2|+\cdots+|x_n|$

$=\sum_{k=1}^n |x_k|$

$一般化されたノルム$

$\lVert\vec{x}\rVert_p = (|x_1|^p+|x_2|^p+\cdots+|x_n|^p)^{\frac{1}{p}}$

$=(\sum_{k=1}^n |x_k|^p)^{\frac{1}{p}}$

```python
# ノルム(ベクトルの大きさ)の実装
import numpy as np
a = np.array([1, 1, -1, -1])
print("--- L2ノルム ---")
print(np.linalg.norm(a)) # L2ノルム（デフォルト）=三平方の定理の斜面の長さ
print("--- L1ノルム ---")
print(np.linalg.norm(a, 1)) # L1ノルム=絶対値の和
```

$逆行列とは$

$たとえばスカラーの場合、3*\frac{1}{3}=1 の \frac{1}{3} が逆数となる$

$行列に掛けると単位行列となる行列を逆行列という$

$行列Aの逆行列をA^-1と表すと、A^-1 A = AA^-1 = E(単位行列) となる$

$逆行列が存在するかは行列式(|A| = \det A = ad-bc(正方行列の場合) \ne 0)により判定することができる$

$逆行列が存在する場合、A^-1 = \frac{1}{ad-bc} \bigl(\begin{smallmatrix} d & -b \\ -c & a \end{smallmatrix}\bigl) で求められる$




```python
# 行列式の実装
# A = 2×2の行列式は、|A| = det A = a*d-b*c
import numpy as np
a = np.array([[1, 2],
              [3, 4]])
print(np.linalg.det(a)) # 行列式が0にならない場合（逆行列が存在する）
b = np.array([[1, 2],
              [0, 0]])
print(np.linalg.det(b)) # 行列式が0になる場合（逆行列が存在しない）
```

```python
# 逆行列の実装
import numpy as np
# 行列
a = np.array([[1, 1],
              [1, 2]])
print(a)
print()
# 逆行列
if np.linalg.det(a) != 0: # 逆行列が存在するか
    b = np.linalg.inv(a) # 逆行列（行列に掛けると単位行列になる）
    print(b)
    print()
    print(np.dot(a, b)) # 単位行列（行列と逆行列の積）
```

$固有値と固有ベクトル$

$A を行列として A\vec{x} = λ\vec{x} とした場合$  
$\vec{x} = 固有ベクトル、λ = 固有値、\det (A - λI) = 0 が固有方程式となる$

$固有ベクトル \vec{x} とは、行列に掛けても向きが変わらず、長さだけが変わるベクトルである$  
$その時に長さが何倍になるかが、対応する固有値 λ となる$

$A\vec{x} = λ\vec{x} → A\vec{x} - λ\vec{x} = \vec{0} → (A - λI)\vec{x} = \vec{0} ... I は単位行列$  
$この式が成り立つとき、行列 A の固有値は λ で、固有ベクトルは \vec{x} である$

$(A - λI)\vec{x} = \vec{0} であり、\vec{x} != \vec{0} ならば、|A - λI| = \vec{0} となる$  
$この式を固有方程式と呼ぶ$

```python
# 固有値と固有ベクトルの計算
import numpy as np
a = np.array([[3, 1],
[2, 4]])
ev = np.linalg.eig(a) # 固有値と固有ベクトルを同時に求める
print(ev[0]) # 最初の要素は固有値
print()
print(ev[1]) # 次の要素は固有ベクトル
```

$コサイン類似度$

$ベクトルの内積は各要素の積の和と等しい$  
$\vec{a}\cdot\vec{b} = a_1 b_1 + a_2 b_2$

$この内積はL2ノルム（ベクトルの長さ）と三角関数で求めることができる$  
$\vec{a}\cdot\vec{b} = \|\vec{a}\|\|\vec{b}\|\cos\theta = \sqrt{a_1^2+a_2^2}\sqrt{b_1^2+b_2^2}\cos\theta$

$これにより \cos\theta は下記となる$  
$\cos\theta = \frac{a_1b_1+a_2b_2}{\sqrt{a_1^2+a_2^2}\sqrt{b_1^2+b_2^2}} = \frac{a_1b_1+a_2b_2}{\|\vec{a}\|\|\vec{b}\|}$  

$\cos\theta の値はベクトル間の角度 \theta が 0 のときに最大値をとり、角度が大きくなると小さくなる$  
$この \cos\theta の値は「2つのベクトルの向きがどれだけ揃っているか」を表す$

```python
# コサイン類似度の計算

import numpy as np

def cos_sim(vec_1, vec_2):
    return np.dot(vec_1, vec_2) / (np.linalg.norm(vec_1) * np.linalg.norm(vec_2))

a = np.array([2, 2, 2, 2])
b = np.array([1, 1, 1, 1]) # aと同じ向き
c = np.array([-1, -1, -1, -1]) # aと反対向き
print("--- aとbのコサイン類似度 ----")
print(cos_sim(a, b))
print("--- aとcのコサイン類似度 ----")
print(cos_sim(a, c))
```

$極限$

$関数における変数の値をある値に近づけるとき、関数の値が限りなく近づく値のこと$
$例えば、関数 y = x^2+1 において、x を限りなく 0 に近づけたとき、 y が限りなく 1 に近づく$
$\lim_{x \to 0} y = \lim_{x \to 0}(x^2+1) = 1$

$f(a)が定数であれば、\lim_{x \to a} f(x) = f(a) は単純代入形となる$
$単純代入できない場合、\lim_{x \to a} f(x) の極限は、\infty や -\infty となることもある$


$微分$

* 微分＝接線の傾きと関数の振る舞い
* ある点で接する接線の傾きが正ならば、その点の近くでグラフは右肩上がり
* ある点で接する接線の傾きが負ならば、その点の近くでグラフは右肩下がり
* 関数が最小値をとる点に接する接線の傾きは0

$関数 y=f(x) において、xの微小な変化量を\Delta xとすると、$
$xを\Delta xだけ変化させた際のyの値は次のようになる$

$y = f(x + \Delta x)$

$y の微小な変化量は下記となる$

$\Delta y = f(x + \Delta x) - f(x)$

$従って、yの微小な変化 \Delta y とxの微小な変化 \Delta x の割合は次の式で表される$

$\frac{\Delta y}{\Delta x} = \frac{f(x+\Delta x)-f(x)}{\Delta x}$

$この式で \Delta x の値を 0 に限りなく近づける極限を f'(x) と表す$

$f'(x) = \frac{df(x)}{dx} = \frac{d}{dx} f(x) =  \lim_{x \to 0} \frac{f(x+\Delta x)-f(x)}{\Delta x}$

$この関数 f'(x) を f(x) の導関数といい、$  
$関数 f(x) から導関数 f'(x) を得ることを関数 f(x) を微分するという$







$微分の公式$

$r を任意の実数として f(x) = x^r としたとき以下が成り立つ$

$\frac{d}{dx} f(x) = \frac{d}{dx} x^r = rx^{r-1}$

$関数の和 f(x) + g(x) の微分$

$\frac{d}{dx} (f(x)+g(x)) = \frac{d}{dx} f(x) + \frac{d}{dx} g(x)$

$関数の積 f(x) g(x) の微分$

$\frac{d}{dx} (f(x)g(x)) = f(x) \frac{d}{dx} g(x) + g(x) \frac{d}{dx} f(x)$

$定数は微分の外に出すことができる$

$\frac{d}{dx} kf(x) = k \frac{d}{dx} f(x)$

```python
# 接線の描画

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return 3*x**2 + 4*x - 5

def my_func_dif(x): # 導関数
    return 6*x + 4

x = np.linspace(-3, 3)
y = my_func(x)
a = 1
# x=1のときの接線。接線の方程式を使用
y_t = my_func_dif(a)*x - my_func_dif(a)*a + my_func(a) 

plt.plot(x, y, label="y")
plt.plot(x, y_t, label="y_t")
plt.legend()
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

$合成関数$

$合成関数とは、y=f(u) u=g(x) のように複数の関数の合成で表される関数のこと$  
$例えば関数 y=(x^2+1)^3 は、以下のような u を挟んだ合成関数と考えることができる$

$y = u^3$  
$u = x^2 + 1$

$連鎖率$

$合成関数の微分は構成する各関数の導関数の積で表すことができ連鎖率という$

$\frac{dy}{dx} = \frac{dy}{du} \frac{du}{dx}$

$y が u の関数で、u が x の関数であるとき、上記公式を用いて y を x で微分することができる$

$例1：y = (x^3 + 2x^2 + 3x +4)^3$

$u = x^3 + 2x^2 + 3x + 4 とする$  
$これにより y = u^3 と表すことができる$

$連鎖率の公式を用いて y を x で微分する$

$\frac{dy}{dx} = \frac{dy}{du} \frac{du}{dx} = 3u^2(3x^2 + 4x + 3)$  
$=3(x^3 + 2x^2 + 3x + 4)^2 (3x^2 + 4x +3)$

$例2：y = (x^2 + 4x + 1)^4$

$u = x^2 + 4x + 1$  
$y = u^4$

$\frac{dy}{dx} = \frac{dy}{du} \frac{du}{dx} = 4u^3(2x + 4)$  
$=4(x^2 + 4x + 1)^3 (2x + 4)$

$偏微分$

$多変数関数を1つの変数により微分することを偏微分という$
$偏微分の場合、他の変数は定数として扱う$

$2変数からなる関数 f(x,y)の偏微分は、 \partial（パーシャル、ディー）の記号で表す$

$\frac{\partial}{\partial x} f(x,y) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x}$

$xのみ微小量 \Delta x だけ変化させて、\Delta x を限りなく0に近づける$  
$yは微小変化しないため、偏微分の際は定数のように扱うことができる$

$偏微分により求めた関数を偏導関数という$

$f(x,y) = 3x^2 + 4xy + 5y^3$

$yを定数として扱い、微分の公式を用いてxで微分する$

$\frac{\partial}{\partial x} f(x,y) = 6x + 4y$

$この場合、偏導関数はyの値を固定した際の、xの変化に対するf(x,y)の変化の割合となる$

$yによる偏微分は以下となる$

$\frac{\partial}{\partial y} f(x,y) = 4x + 15y^2$

$これは、xの値を固定した際の、yの変化に対するf(x,y)の変化の割合となる$

$例1：f(x,y) = 2x^3 + 4x^2y + xy^2 - 4y^2$

$\frac{\partial}{\partial x}(x,y) = 6x^2 + 8xy + y^2 \cdots xで（yを固定して）偏微分$

$\frac{\partial}{\partial y}(x,y) = 4x^2 + 2xy + 8y \cdots yで（xを固定して）偏微分$

$全微分$

$多変数関数の微小変化を、すべての変数の微小変化を使って求めることを全微分という$  
$全微分は、偏微分により求めた勾配を使って求める$

$2変数関数z = f(x,y)の全微分は次の式で表される$

$dz = \frac{\partial z}{\partial x} dx + \frac{\partial z}{\partial y} dy$

$xによる偏微分にxの微小変化dxを掛けたものと、yによる偏微分にyの微小変化dyを$  
$掛けたものを足し合わせてzの微小変化dzとする$

$n個の変数を持つ関数zの全微分(x_iが各変数を表す)$

$dz = \sum_{i=1}^n \frac{\partial z}{\partial x_i} dx_i$

$全微分を用いることで、多変数関数の微小な変化量を各変数による偏微分と$  
$各変数の微小な変化により求めることができる$

$例1：f(x,y) = 3x^2 + 4xy + 5y^3$

$xによる偏微分$  
$\frac{\partial f}{\partial x} = 6x + 4y$

$yによる偏微分$  
$\frac{\partial f}{\partial y} = 4x + 15y^2$

$全微分$  
$dz = (6x + 4y)dx + (4x + 15y^2)dy$

$多変数合成関数の連鎖律$

$多変数合成関数の微分は、構成する各関数の偏微分の積で表すことができる$

$z = f(u,v)$

$u = g(x)$

$v = h(x)$

$zはuとvの関数で、uとvはそれぞれxの関数であり、この合成関数で \frac{dz}{dx} を求める$

$dz = \frac{\partial z}{\partial u} du + \frac{\partial z}{\partial v} dv$

$この式の両辺を微小量dxで割ることで、合成関数zのxによる微分を次の式のように得ることができる$

$\frac{dz}{dx} = \frac{\partial z}{\partial u} \frac{du}{dx} + \frac{\partial z}{\partial v} \frac{dv}{dx}$

$この式を一般化する。uやvのような媒介する変数がm個あるとすると、次のように表すことができる$

$\frac{dz}{dx} = \sum^m_{i=1} \frac{\partial z}{\partial u_i} \frac{\partial u_i}{\partial x}$
